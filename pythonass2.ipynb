2. Explain the difference between mutable and immutable data types with examples

    Mutable data types can be modified after creation whereas immutable datatypes cannot be modified after creation.In mutable data types changes in value occur in the same memory location whereas in immutable data types a new memory location is assigned for modified values.
    Example of Mutable data type:
    my_list = [1, 2, 3]
    print(id(my_list))  # Memory address before modification
    my_list.append(4)  # Modifying the list
    print(my_list)  # Output: [1, 2, 3, 4]
    print(id(my_list))  # Memory address remains the same
    Example of immutable data type:
    my_string = "Hello"
    print(id(my_string))  # Memory address before modification
    my_string = my_string + " World"  # Modifying the string
    print(my_string)  # Output: "Hello World"
    print(id(my_string))  # Memory address changes
